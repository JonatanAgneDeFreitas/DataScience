1. Separe a base em treinamento e teste (25% para teste, 75% para treinamento).

```python
df_treino, df_teste = train_test_split(df, test_size=0.25, random_state=42)
```

```python
modelo = '''renda ~ sexo 
                    + posse_de_veiculo 
                    + posse_de_imovel 
                    + qtd_filhos 
                    + tipo_renda 
                    + educacao 
                    + estado_civil 
                    + tipo_residencia 
                    + idade 
                    + tempo_emprego 
                    + qt_pessoas_residencia
                    '''
y_treino, X_treino = patsy.dmatrices(formula_like=modelo, data=df_treino)
y_teste, X_teste = patsy.dmatrices(formula_like=modelo, data=df_teste)
```

2. Rode uma regulariza√ß√£o ridge com alpha = [0, 0.001, 0.005, 0.01, 0.05, 0.1] e avalie o  ùëÖ2
  na base de testes. Qual o melhor modelo?

```python
alphas = [0, 0.001, 0.005, 0.01, 0.05, 0.1]
r2_y_pred = []

for alpha in alphas:
    md = sm.OLS(y_treino, X_treino)
    reg = md.fit_regularized(method='elastic_net', 
                             refit=True, 
                             L1_wt=0,
                             alpha=alpha)
    y_pred = reg.predict(X_teste)
    r2 = r2_score(y_teste, y_pred)
    
    r2_y_pred.append(r2)
    print(f'Alpha {alpha}: \nR-squared = {r2}\n')
    
pd.DataFrame({'alpha':alphas, 
              'ùëÖ2':r2_y_pred
             }).sort_values(by='ùëÖ2', ascending=False)
```

3. Fa√ßa o mesmo que no passo 2, com uma regress√£o LASSO. Qual m√©todo chega a um melhor resultado?

```python
alphas = [0, 0.001, 0.005, 0.01, 0.05, 0.1]
r2_y_pred = []

for alpha in alphas:
    md = sm.OLS(y_treino, X_treino)
    reg = md.fit_regularized(method='elastic_net', 
                             refit=True, 
                             L1_wt=1,
                             alpha=alpha)
    y_pred = reg.predict(X_teste)
    r2 = r2_score(y_teste, y_pred)
    
    r2_y_pred.append(r2)
    print(f'Alpha {alpha}: \nR-squared = {r2}\n')
    
pd.DataFrame({'alpha':alphas, 
              'ùëÖ2':r2_y_pred
             }).sort_values(by='ùëÖ2', ascending=False)
```

4. Rode um modelo stepwise. Avalie o ùëÖ2 na vase de testes. Qual o melhor resultado?

```python
df_dummies = pd.get_dummies(data=df.dropna(), drop_first=True).astype(int)

X = df_dummies.drop(columns='renda')
y = df_dummies['renda']

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
def stepwise_selection(X, y, 
                       initial_list=[], 
                       threshold_in=0.05, 
                       threshold_out = 0.05, 
                       verbose=True):

    included = list(initial_list)
    while True:
        changed=False
        
        excluded = list(set(X.columns)-set(included))
        new_pval = pd.Series(index=excluded, dtype=np.dtype('float64'))
        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included+[new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min()
        if best_pval < threshold_in:
            best_feature = new_pval.index[new_pval.argmin()]
            included.append(best_feature)
            changed=True
            if verbose:
                 print('Add  {:30} with p-value {:.6}'.format(best_feature, best_pval))

        print("#############")
        print(included)
        model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()

        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max()
        if worst_pval > threshold_out:
            changed=True
            worst_feature = pvalues.argmax()
            included.remove(worst_feature)
            if verbose:
                print('Drop {:30} with p-value {:.6}'.format(worst_feature, worst_pval))
        if not changed:
            break
    return included

variaveis = stepwise_selection(X_teste, y_teste)

print('\n\nresulting features:')
print(variaveis)
```

```python
X = df_dummies[variaveis]
y = df_dummies['renda']
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.25, random_state=42)

y_pred = sm.OLS(y_treino, X_treino).fit().predict(X_teste)
print(f'R-squared: {r2_score(y_teste, y_pred)}')
```

```python
alphas = [0, 0.001, 0.005, 0.01, 0.05, 0.1]
r2_y_pred = []

for alpha in alphas:
    md = sm.OLS(y_treino, X_treino)
    reg = md.fit_regularized(method='elastic_net', 
                             refit=True, 
                             L1_wt=0,
                             alpha=alpha)
    y_pred = reg.predict(X_teste)
    r2 = r2_score(y_teste, y_pred)
    
    r2_y_pred.append(r2)
    print(f'Alpha {alpha}: \nR-squared = {r2}\n')
    
pd.DataFrame({'alpha':alphas, 
              'ùëÖ2':r2_y_pred
             }).sort_values(by='ùëÖ2', ascending=False)
```

```python
alphas = [0, 0.001, 0.005, 0.01, 0.05, 0.1]
r2_y_pred = []

for alpha in alphas:
    md = sm.OLS(y_treino, X_treino)
    reg = md.fit_regularized(method='elastic_net', 
                             refit=True, 
                             L1_wt=1,
                             alpha=alpha)
    y_pred = reg.predict(X_teste)
    r2 = r2_score(y_teste, y_pred)
    
    r2_y_pred.append(r2)
    print(f'Alpha {alpha}: \nR-squared = {r2}\n')
    
pd.DataFrame({'alpha':alphas, 
              'ùëÖ2':r2_y_pred
             }).sort_values(by='ùëÖ2', ascending=False)
```

O melhor resultado dos modelos stepwise foi o que utilizou a regress√£o LASSO, pois obteve desempenho semelhante em todos os valores de alpha e ainda assim alcan√ßou um coeficiente de determina√ß√£o compar√°vel ao modelo que utilizou a regulariza√ß√£o ridge com o maior valor desse coeficiente. Esse modelo pode ser considerado mais simples e, ao mesmo tempo, apresentar o melhor resultado entre as op√ß√µes desenvolvidas.

6. Partindo dos modelos que voc√™ ajustou, tente melhorar o  ùëÖ2 na base de testes. Use a criatividade, veja se consegue inserir alguma transforma√ß√£o ou combina√ß√£o de vari√°veis.

```python
df_matrix = df_dummies.loc[:, variaveis + ['renda']]
df_matrix.rename(columns={'educacao_Superior completo': 'educacao_Superior_completo'}, 
                 inplace=True)

df_treino, df_teste = train_test_split(df_matrix, test_size=0.25, random_state=42)

modelo = smf.ols(formula='''
                         np.log(renda) ~ tempo_emprego 
                                         + sexo_M 
                                         + educacao_Superior_completo
                                         + idade 
                                         - posse_de_veiculo 
                                         + tipo_renda_Empres√°rio 
                                         + qt_pessoas_residencia
                                         - tipo_residencia_Est√∫dio
                                         ''', data=df_treino)

reg = modelo.fit_regularized(method='elastic_net', 
                             refit=True, 
                             L1_wt=1,
                             alpha=0)

print(reg.summary())

y_pred = reg.predict(df_teste)
print('\nCoeficiente de determina√ß√£o (ùëÖ2) na base de testes:', 
      r2_score(df_teste['renda'], 
               np.exp(y_pred)
              )
     )
```
