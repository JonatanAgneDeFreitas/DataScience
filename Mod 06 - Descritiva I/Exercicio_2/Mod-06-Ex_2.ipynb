```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", message="use_inf_as_na option is deprecated", category=FutureWarning)

%matplotlib inline

dados = pd.read_csv('./input/dados.csv')
dados.replace([float('inf'), float('-inf')], pd.NA, inplace=True)
sns.set_style("whitegrid")
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=dados, y='IDADEMAE', color='lightblue')
plt.title('Boxplot idade mãe')
plt.ylabel('idade')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=dados, y='PESO', color='lightgreen')
plt.title('Boxplot idade do bebê')
plt.ylabel('Peso (g)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=dados, x='IDADEMAE', bins=20, kde=True, color='blue')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=dados, x='PESO', bins=20, kde=True, color='green')
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso (gramas)')
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.5)
plt.show()
```
