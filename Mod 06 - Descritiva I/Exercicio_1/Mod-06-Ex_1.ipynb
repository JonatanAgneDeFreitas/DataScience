#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

dados = pd.read_csv('./Input/dados.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.figure(figsize=(10, 6))
dados['APGAR1'].value_counts().sort_index().plot(kind='bar')
plt.title('APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
dados['APGAR5'].value_counts().sort_index().plot(kind='bar')
plt.title('APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
dados['SEXO'].value_counts().sort_index().plot(kind='bar')
plt.title('SEXO')
plt.xlabel('SEXO')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
dados['GESTACAO'].value_counts().sort_index().plot(kind='bar')
plt.title('GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.show()
```

```python
map_estados_civis = {
    1: 'Solteira',
    2: 'Casada',
    3: 'Viúva',
    4: 'Separado',
    5: 'União Consensual',
    9: 'Ignorado'
}

df_mae = dados.copy()
df_mae['ESTCIVMAE'] = dados['ESTCIVMAE'].map(map_estados_civis)

plt.figure(figsize=(10, 6))
df_mae['ESTCIVMAE'].value_counts().sort_index().plot(kind='bar')
plt.title('ESTADO CIVIL MÃE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apg5_por_consultas = dados.groupby('CONSULTAS')['APGAR5'].mean()
plt.figure(figsize=(10, 6))
apg5_por_consultas.plot(kind='bar')
plt.title('APGAR5 Médio por Consultas')
plt.xlabel('Consultas')
plt.ylabel('APGAR5 Médio')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

dados['SEXO'].value_counts().sort_index().plot(kind='bar', ax=axs[0, 0])
axs[0, 0].set_title('SEXO')
axs[0, 0].set_xlabel('SEXO')
axs[0, 0].set_ylabel('Frequência')

dados['GESTACAO'].value_counts().sort_index().plot(kind='bar', ax=axs[0, 1])
axs[0, 1].set_title('GESTACAO')
axs[0, 1].set_xlabel('GESTACAO')
axs[0, 1].set_ylabel('Frequência')

df_mae['ESTCIVMAE'].value_counts().sort_index().plot(kind='bar', ax=axs[1, 0])
axs[1, 0].set_title('ESTADO CIVIL MÃE')
axs[1, 0].set_xlabel('ESTCIVMAE')
axs[1, 0].set_ylabel('Frequência')

dados['CONSULTAS'].value_counts().sort_index().plot(kind='bar', ax=axs[1, 1])
axs[1, 1].set_title('CONSULTAS PRÉ NATAL')
axs[1, 1].set_xlabel('CONSULTAS')
axs[1, 1].set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```
